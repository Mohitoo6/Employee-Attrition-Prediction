```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Load the dataset
file_path = "C:/Users/rabik/OneDrive/Desktop/Data Science for Business Package/1. Human Resources Data/Human_Resources.csv"
df = pd.read_csv(file_path)

# Display dataset information
print("Dataset Information:")
df.info()
print("\nFirst 5 rows:")
print(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Encode categorical variables before correlation analysis
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Attrition distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=df, palette='coolwarm')
plt.title("Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()

# Visualizing numerical feature distributions
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
fig, axes = plt.subplots(len(numerical_features) // 3 + 1, 3, figsize=(20, 15))
axes = axes.flatten()
for i, col in enumerate(numerical_features):
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i], color='blue')
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# Boxplot for salary-related features
plt.figure(figsize=(8, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='coolwarm')
plt.title("Monthly Income vs Attrition")
plt.show()

# Relationship between Age and Attrition
plt.figure(figsize=(8, 6))
sns.boxplot(x='Attrition', y='Age', data=df, palette='coolwarm')
plt.title("Age vs Attrition")
plt.show()

# Relationship between Job Role and Attrition
plt.figure(figsize=(12, 6))
sns.countplot(y='JobRole', hue='Attrition', data=df, palette='coolwarm')
plt.title("Attrition by Job Role")
plt.show()

# Relationship between Department and Attrition
plt.figure(figsize=(8, 6))
sns.countplot(y='Department', hue='Attrition', data=df, palette='coolwarm')
plt.title("Attrition by Department")
plt.show()

# Relationship between Overtime and Attrition
plt.figure(figsize=(8, 6))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='coolwarm')
plt.title("Attrition by OverTime")
plt.show()

# New Visualizations
# Pairplot for selected numerical features
selected_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'TotalWorkingYears', 'Attrition']
sns.pairplot(df[selected_features], hue='Attrition', palette='coolwarm')
plt.show()

# Distribution of EducationField with Attrition
plt.figure(figsize=(10, 6))
sns.countplot(y='EducationField', hue='Attrition', data=df, palette='coolwarm')
plt.title("Attrition by Education Field")
plt.show()

# Swarmplot to visualize YearsAtCompany vs MonthlyIncome
plt.figure(figsize=(12, 6))
sns.swarmplot(x='YearsAtCompany', y='MonthlyIncome', hue='Attrition', data=df, palette='coolwarm', alpha=0.7)
plt.title("Years at Company vs Monthly Income by Attrition")
plt.show()

# Violin plot to show distribution of TotalWorkingYears
plt.figure(figsize=(10, 6))
sns.violinplot(x='Attrition', y='TotalWorkingYears', data=df, palette='coolwarm')
plt.title("Distribution of Total Working Years by Attrition")
plt.show()
```
